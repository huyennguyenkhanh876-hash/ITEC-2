```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from google.colab import drive

# 1. Kết nối Drive
drive.mount('/content/drive', force_remount=True)

# 2. Cấu hình đường dẫn (Đã sửa theo ảnh của bạn)
folder_path = '/content/drive/MyDrive/ITEC 2/.vnstock'
vn30_2026 = [
    'MWG', 'HPG', 'SAB', 'BID', 'VPL', 'PLX', 'CTG', 'GVR', 'VHM', 'VCB',
    'SSB', 'BCM', 'VIB', 'HDB', 'GAS', 'MSN', 'FPT', 'ACB', 'TPB', 'VIC',
    'SHB', 'VRE', 'MBB', 'TCB', 'VPB', 'VJC', 'SSI', 'DGC', 'VNM', 'LPB', 'STB'
]

features = ['open', 'high', 'low', 'close', 'volume']
pdf_filename = 'Bao_Cao_SVR_VN30.pdf'

def run_svr_analysis_direct_print():
    summary_list = []

    # Mở file PDF để lưu song song
    with PdfPages(pdf_filename) as pdf:
        for symbol in vn30_2026:
            file_path = os.path.join(folder_path, f"stock_data_{symbol}.csv")

            if os.path.exists(file_path):
                try:
                    df = pd.read_csv(file_path)
                    df['time'] = pd.to_datetime(df['time'])
                    df = df.sort_values('time')

                    X_raw = df[features]
                    y_raw = df['close'].shift(-1)

                    X_raw = X_raw[:-1]
                    y_raw = y_raw[:-1]

                    split = int(len(X_raw) * 0.8)
                    X_train_raw, X_test_raw = X_raw[:split], X_raw[split:]
                    y_train, y_test = y_raw[:split], y_raw[split:]
                    test_dates = df['time'].iloc[split:-1]

                    # Chuẩn hóa
                    scaler_X = StandardScaler()
                    X_train = scaler_X.fit_transform(X_train_raw)
                    X_test = scaler_X.transform(X_test_raw)

                    # Huấn luyện SVR
                    model = SVR(kernel='rbf', C=1000, gamma=0.1)
                    model.fit(X_train, y_train)
                    y_pred = model.predict(X_test)

                    # Tính chỉ số
                    mse = mean_squared_error(y_test, y_pred)
                    mae = mean_absolute_error(y_test, y_pred)
                    r2 = r2_score(y_test, y_pred)
                    summary_list.append({'Mã': symbol, 'MSE (Loss)': mse, 'MAE': mae, 'R2': r2})

                    # --- IN KẾT QUẢ VÀ BIỂU ĐỒ TRỰC TIẾP ---
                    print(f"\n>>> Đang hiển thị kết quả mã: {symbol}")
                    print(f"Loss (MSE): {mse:.2f} | MAE: {mae:.2f} | R2: {r2:.4f}")

                    plt.figure(figsize=(10, 4))
                    plt.plot(test_dates, y_test.values, label='Thực tế', color='#1f77b4')
                    plt.plot(test_dates, y_pred, label='Dự báo SVR', color='#ff7f0e', linestyle='--')
                    plt.title(f"Dự báo {symbol} - SVR Model")
                    plt.legend()
                    plt.grid(True, alpha=0.3)

                    # Lưu vào PDF
                    pdf.savefig()
                    # Hiển thị trực tiếp ra màn hình Colab
                    plt.show()
                    plt.close()

                except Exception as e:
                    print(f"Lỗi tại mã {symbol}: {e}")
            else:
                pass # Bỏ qua nếu không thấy file

    # --- IN BẢNG TỔNG HỢP CUỐI CÙNG ---
    print("\n" + "="*50)
    print("BẢNG TỔNG HỢP HIỆU SUẤT TẤT CẢ CÁC MÃ")
    print("="*50)
    final_df = pd.DataFrame(summary_list)
    # Sắp xếp theo R2 để xem mã nào dự báo tốt nhất
    print(final_df.sort_values(by='R2', ascending=False).to_string(index=False))

# Chạy trực tiếp
run_svr_analysis_direct_print()
```
