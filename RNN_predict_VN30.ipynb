```python
import os
import glob
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Cố định random seed để kết quả có thể tái lập
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Danh sách VN30 (Áp dụng từ 02/02/2026)
VN30_TICKERS = [
    'ACB', 'BID', 'CTG', 'DGC', 'FPT', 'GAS', 'GVR', 'HDB', 'HPG', 'LPB',
    'MBB', 'MSN', 'MWG', 'PLX', 'SAB', 'SHB', 'SSB', 'SSI', 'STB', 'TCB',
    'TPB', 'VCB', 'VHM', 'VIB', 'VIC', 'VJC', 'VNM', 'VPB', 'VPL', 'VRE'
]

def get_vn30_files(dataset_folder):
    """Quét thư mục và trả về danh sách các tuple (mã_cổ_phiếu, đường_dẫn_file) thuộc VN30."""
    search_pattern = os.path.join(dataset_folder, 'stock_data_*.csv')
    all_files = glob.glob(search_pattern)

    vn30_files_info = []
    for file_path in all_files:
        filename = os.path.basename(file_path)
        ticker = filename[11:-4].upper()
        if ticker in VN30_TICKERS:
            vn30_files_info.append((ticker, file_path))

    return vn30_files_info
```

```python
# ==========================================
# THIẾT LẬP THAM SỐ CHUNG
# ==========================================
folder_path = '/content/drive/MyDrive/RNNStock/.vnstock'  # Thay bằng đường dẫn tới thư mục chứa file CSV của bạn
features = ['open', 'close', 'high', 'low', 'volume']
seq_length = 1
input_dim = len(features)
hidden_dim = 32
output_dim = 1
num_epochs = 30    # Số vòng lặp huấn luyện cho mỗi mã
batch_size = 64

print("Đang quét các file dữ liệu...")
vn30_files = get_vn30_files(folder_path)
print(f"Đã tìm thấy {len(vn30_files)} file thuộc danh mục VN30.\n")

if len(vn30_files) == 0:
    raise FileNotFoundError("Không tìm thấy dữ liệu. Vui lòng kiểm tra lại 'folder_path'.")
```

```python
# ==========================================
# VÒNG LẶP XỬ LÝ VÀ HUẤN LUYỆN TỪNG FILE
# ==========================================
for ticker, file_path in sorted(vn30_files):
    print("="*70)
    print(f"Đang xử lý và huấn luyện mã cổ phiếu: {ticker} (TensorFlow)")
    print("="*70)

    # 1. Đọc và tiền xử lý dữ liệu
    df = pd.read_csv(file_path)
    df['time'] = pd.to_datetime(df['time'])
    df = df.sort_values('time').reset_index(drop=True)
    df['target'] = df['open'].shift(-1)
    df.dropna(inplace=True)

    # 2. Chuẩn hóa dữ liệu (Scale riêng cho từng mã)
    feature_scaler = MinMaxScaler(feature_range=(0, 1))
    target_scaler = MinMaxScaler(feature_range=(0, 1))

    X_scaled = feature_scaler.fit_transform(df[features].values)
    y_scaled = target_scaler.fit_transform(df[['target']].values)

    X = X_scaled.reshape(-1, seq_length, input_dim)
    y = y_scaled

    # 3. Chia tập Train/Test (80% / 20%)
    train_size = int(len(X) * 0.8)

    X_train, y_train = X[:train_size], y[:train_size]
    X_test, y_test = X[train_size:], y[train_size:]

    # Keras nhận thẳng Numpy Array nên ta KHÔNG cần chuyển sang Tensor như PyTorch

    # 4. Khởi tạo mô hình MỚI cho từng mã cổ phiếu (Sử dụng Keras Sequential API)
    model = Sequential([
        SimpleRNN(hidden_dim, activation='relu', input_shape=(seq_length, input_dim)),
        Dense(output_dim)
    ])

    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    # 5. Huấn luyện mô hình
    # verbose=0 để ẩn bớt log chạy chi tiết từng epoch của Keras cho đỡ rối mắt
    model.fit(X_train, y_train, batch_size=batch_size, epochs=num_epochs, verbose=0)

    # 6. Đánh giá (Tính Loss trên Train và Test)
    train_loss = model.evaluate(X_train, y_train, verbose=0)
    test_loss = model.evaluate(X_test, y_test, verbose=0)

    print(f"-> Train Loss (MSE): {train_loss:.6f}")
    print(f"-> Test Loss  (MSE): {test_loss:.6f}")

    # Lấy mảng dự đoán cho tập Test
    test_pred = model.predict(X_test, verbose=0)

    # 7. Nghịch đảo Scaler để lấy giá trị thực vẽ biểu đồ
    y_test_real = target_scaler.inverse_transform(y_test)
    test_pred_real = target_scaler.inverse_transform(test_pred)

    # 8. Vẽ biểu đồ y_true vs y_predicted
    plt.figure(figsize=(12, 5))
    plt.plot(y_test_real, label='Giá Thực Tế (y_true)', color='blue', alpha=0.7)
    plt.plot(test_pred_real, label='Giá Dự Đoán (y_predicted)', color='red', alpha=0.7)
    plt.title(f'Biểu đồ Dự Đoán Giá Mở Cửa (Tập Test) - Mã: {ticker}')
    plt.xlabel('Thời gian (Index tập Test)')
    plt.ylabel('Giá Cổ Phiếu')
    plt.legend()
    plt.grid(True)

    # Hiển thị biểu đồ cho file hiện tại. Sau khi render xong nó mới chạy vòng lặp sang file tiếp theo.
    plt.show()
    print("\n") # Xuống dòng cho dễ nhìn trước khi sang mã khác

print("="*70)
print("Đã hoàn tất quá trình huấn luyện bằng TensorFlow cho tất cả các mã VN30!")
```
